```python
import nltk
from nltk.corpus import gutenberg
ausContent=gutenberg.words('austen-persuasion.txt')
print(ausContent)

for fileid in gutenberg.fileids():
    print(fileid)
    
ausWords=nltk.corpus.gutenberg.words('austen-persuasion.txt')
noOfWords=len(ausContent)
print('The number of wrods is' , + noOfWords)
uniqueWords=set(ausContent)
print(uniqueWords)

characters=list(ausContent)
d=dict()
for c in characters:
    if c not in d:
        d[c]=1
    else:
        d[c]+=1
            
sorted_characters= sorted(d.items(),key=lambda ausContent : ausWords[1],reverse=True)
print('Characters and teir frequency is :' +str(sorted_characters))

top10=sorted_characters[:10]
print('The top 10 most frequent words and their frequency is' + str(top10))
```
