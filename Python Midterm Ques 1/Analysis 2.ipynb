# Analysis 2- Finding out frequency of emails based on the domain which they belong to and plotting a bar graph

```python
#necessary libraries are imported
import email
import os
import functools
import itertools
import matplotlib.pyplot as plt
import operator
```

```python
path="C:\\Users\\smrut\\midterm\\data\\enron\\enron_mail_20150507\\maildir"
root=path                                                                            
print(root)
```

```python
# Checking the To field of the email and splitting it to find the 2nd part of the email address ie the domain.
#If the domain is present in the sent mails then it is added to the datatoanalyse dictionary
def addtodictionary(raw_email):
    email_message = email.message_from_string(raw_email)
    for to in str(email_message['To']).split(','):
        if to.endswith("@enron.com") is False and to != "None":
            k = to.split("@",1)[1]
            if k in datatoanalyse:
                datatoanalyse[k] +=1
            else:
                datatoanalyse[k] = 1
```

```python
datatoanalyse = {}
for subdir, dirs, files in os.walk(root):
    if subdir.endswith("_sent_mail"):
        for file in files:
            with open(subdir+"/"+file, 'r') as f:
                addtodictionary(f.read())
```

```python
#frequency of emails under each website is calculated.
freq = sorted(datatoanalyse.items(), key=operator.itemgetter(1), reverse=True)[:10]
freq.reverse()
labels, values = zip(*freq)
```

```python
plt.barh(range(len(values)), values, align='center')
plt.yticks(range(len(labels)), labels)
fig = plt.gcf()
fig.set_size_inches(15, 10)
plt.suptitle('Website Wise Frequency of emails' , fontsize=20, fontweight='bold')
plt.ylabel("Website Name", fontsize=15)
plt.xlabel("No of emails", fontsize=15)
plt.show()
```
