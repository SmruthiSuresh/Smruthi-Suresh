# Analysis1- Analysis is done on a particular university to show the score vs the world rank relationship and also a bar graph is plotted to show the evolution of research,income etc through 2011 to 2016


```python
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python

# times data is read from the data folder
timesData = pd.read_csv('data/timesData.csv')

# eplf rankings are taken into consideration
times_epfl_df = timesData[timesData['university_name'] == ('École Polytechnique Fédérale de Lausanne')] 


```

```python
# Converting int into floats
times_epfl_df['world_rank'] = times_epfl_df['world_rank'].astype('int64')
times_epfl_df['international'] = times_epfl_df['international'].astype('float64')
times_epfl_df['income'] = times_epfl_df['income'].astype('float64')
times_epfl_df['total_score'] = times_epfl_df['total_score'].astype('float64')

# The evolution of ranking through the different years is plotted.
ax = times_epfl_df.plot(
    kind='line', 
    x='year', 
    y='world_rank',
    xlim=(2011, 2016), 
    ylim=(1, 60), 
    xticks=range(2011, 2017)
)

```

```python
# Checking ranking vs the total score in all the years.
compare_df = times_epfl_df[['total_score', 'world_rank']]

percent_df = compare_df.pct_change()
percent_df = percent_df.dropna()
percent_df['year'] = times_epfl_df['year']

# Score and rank are inversely propotional in the sense that if one goes up the other goes down.
ax = percent_df.plot(
    kind='line', 
    x='year', 
    xticks=range(2012, 2017),# hacky
    ylim=(-1, 1), 
)

ax.set_xticklabels(['2011-2012', '2012-2013', '2013-2014', '2014-2015', '2015-2016'])

percent_df
```

```python
# Plotting a bar graph to show the quality of teaching,research, citations etc from 2011 to 2016
bar_criteria_df = times_epfl_df[['year', 'teaching', 'international', 'research', 'citations', 'income']]
ax = bar_criteria_df.plot.bar(
    stacked=True,
    x='year'
)

ax.legend(loc=7, bbox_to_anchor=(1.4, 0.5)) 
```

```python
# using seaborn module to plot the score against research, income, citations and teaching
sns.lmplot(x="international", y="total_score", data=times_epfl_df)

sns.lmplot(x="research", y="total_score", data=times_epfl_df)
sns.lmplot(x="income", y="total_score", data=times_epfl_df)
sns.lmplot(x="citations", y="total_score", data=times_epfl_df)
sns.lmplot(x="teaching", y="total_score", data=times_epfl_df)
```
