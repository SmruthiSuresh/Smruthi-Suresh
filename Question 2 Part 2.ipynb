```python
dfb=dfa.rename(columns = {'Total Compensation':'Total_Compensation','Total benefits':'Total_Benefits','Other Benefits' : 'Other_Benefits','Total Salary':'Total_Salary','Employee Identifier':'Employee_Identifier','Total Benefits':'Total_Benefits','Year Type':'Year_Type'})

dfb.head()
```

```python
dfb=dfb.loc[dfb.Year_Type=="Calendar"]
dfb.head()
```

```python
dfb['Mean Total_Compensation']=pd.DataFrame(dfb.groupby('Employee_Identifier').Total_Compensation.mean())
dfb['Mean Total_Benefits']=pd.DataFrame(dfb.groupby('Employee_Identifier').Total_Benefits.mean())
dfb['Mean Total_Salary']=pd.DataFrame(dfb.groupby('Employee_Identifier').Total_Salary.mean())
dfb['Mean Other_Benefits']=pd.DataFrame(dfb.groupby('Employee_Identifier').Other_Benefits.mean())
dfb['Mean Total_Compensation']=pd.DataFrame(dfb.groupby('Employee_Identifier').Total_Compensation.mean())
dfb.head()
```

```python
dfb['PERCENT TOTAL BENEFITS'] = (dfb['Total_Salary'] / dfb['Total_Compensation']) *100
dfb.head()
```
