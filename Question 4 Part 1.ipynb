```python
dfa=pd.read_csv('movies_awards.csv')
dfa = dfa.fillna(value=False)
```

```python
dfa['Awards_Won']= dfa['Awards'].str.extract('(\d) win', expand=True)
dfa['Awards_Nominated']=dfa['Awards'].str.extract('(\d) nomination', expand=True)
dfa['Prime_Awards_Won ']=dfa['Awards'].str.extract('Won (\d) Primetime', expand=True)
dfa['Prime_Awards_Nominated ']=dfa['Awards'].str.extract('Nominated for (\d) Primetime', expand=True)
dfa['Oscar_Awards_Won ']=dfa['Awards'].str.extract('Won (\d) Oscar', expand=True)
dfa['Oscar_Awards_Nominated ']=dfa['Awards'].str.extract('Nominated for (\d) Oscar', expand=True)
dfa['BAFTA_Awards_Nominated ']=dfa['Awards'].str.extract('Nominated for (\d) BAFTA Film', expand=True)
dfa['Golden_Globe_Nominated ']=dfa['Awards'].str.extract('Nominated for (\d) Golden Globe', expand=True)
dfa['Golden_Globe_Won ']=dfa['Awards'].str.extract('Won (\d) Golden Globe', expand=True)
```
