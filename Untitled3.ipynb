```python
import csv
import pandas as pd
import re

from pandas import DataFrame
import numpy as np
```

```python
dfa=pd.read_csv('movies_awards.csv')
dfa = dfa.fillna(value=0)
dfa.head()
```

```python
dfa['Awards Won']= dfa['Awards'].str.extract('(\d) win', expand=True)
dfa['Awards Nominated ']=dfa['Awards'].str.extract('(\d) nomination', expand=True).reset_index()
dfa['Prime_Awards_Won ']=dfa['Awards'].str.extract('Won (\d) Primetime', expand=True).reset_index()
dfa['Prime_Awards_Nominated ']=dfa['Awards'].str.extract('Nominated for (\d) Primetime', expand=True).reset_index()
dfa['Oscar_Awards_Won ']=dfa['Awards'].str.extract('Won (\d) Oscar', expand=True).reset_index()
dfa['Oscar_Awards_Nominated ']=dfa['Awards'].str.extract('Nominated for (\d) Oscar', expand=True).reset_index()


dfa['BAFTA_Awards_Nominated ']=dfa['Awards'].str.extract('Nominated for (\d) BAFTA Film', expand=True).reset_index()
dfa['Golden_Globe_Nominated ']=dfa['Awards'].str.extract('Nominated for (\d) Golden Globe', expand=True).reset_index()
dfa['Golden_Globe_Won ']=dfa['Awards'].str.extract('Won (\d) Golden Globe', expand=True).reset_index()
#dfa['Total Awards Won or Nominated'] = dfa.apply(lambda row: row['Awards Won'] + row['Awards Nominated'], axis=1)
dfa
```
